# Adjusting the datasets

```python
a=ratings['user_id'].value_counts()>200
# these are only useful
```

```python
y=a[a].index
```

```python
y
#only useful indexes with > 200 ratings
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
ratings
```

# Combining the datasets

```python
ratings_and_books=ratings.merge(books,on='ISBN')
```

```python
ratings_and_books
```

```python
number_ratings=ratings_and_books.groupby('title')['rating'].count().reset_index()
```

```python
number_ratings
```

```python
number_ratings.rename(columns={'rating':'No of rating'},inplace=True)
```

```python
Rating=ratings_and_books.merge(number_ratings,on='title')
```

```python
Rating
```

```python
final_rating=Rating[Rating['No of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

# Creating a pivot table

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

#  Implementing the Nearest neighbour algorithm

```python
# we will use sparse matrix to eleminate 0 present in pivot and reduce computation time
# it consider only one 
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
NN=NearestNeighbors(algorithm='brute')
# we use brute force here in this algo
```

```python
# training the dataset
NN.fit(book_sparse)
```

# Testing the Model

```python
print(book_pivot.index[237])
```

```python
distances,suggestions=NN.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1))
# 2 arrays are returned
```

```python
distances
# the 5 books nearest
```

```python
suggestions
#id if 5 nearest books
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

# For user

```python
def book_rec(name):
    id=np.where(book_pivot.index==name)[0][0]
    distances,suggestions=NN.kneighbors(book_pivot.iloc[id,:].values.reshape(1,-1))
    print("Suggestion for",name,'are as folows')
    for i in range(len(suggestions)):
        print(book_pivot.index[suggestions[i]])
    
```
