# Pipeline Creation to do above three tasks

```python
# pipeline to remove punctuations and stopwords
def msg_cleaning(msg):
    text_punc_rem=[c for c in msg if c not in string.punctuation]
    text_punc_rem_join="".join(text_punc_rem)
    text_punc_rem_join_clean=[wrd for wrd in punc_rem_join.split( ) if wrd.lower() not in stopwords.words('english')]
    return  text_punc_rem_join_clean
```

```python
tweets_df_clean=tweets_df['tweet'].apply(msg_cleaning)
```

```python
tweets_df_clean[10]
```

```python
tweets_df['tweet'][10]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer(analyzer=msg_cleaning)
tweets_countvectorizer=CountVectorizer(analyzer=msg_cleaning,dtype='uint8').fit_transform(tweets_df['tweet']).toarray()
```

```python
tweets_countvectorizer.shape
```

```python
X=tweets_countvectorizer
```

```python
Y=tweets_df['label']
```

# Train a Naive Bayes Classifier Model

```python
#we know about naive bayes classifier as done in readme.txt
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB_classifier=MultinomialNB()
NB_classifier.fit(X_train,Y_train)
```

# Accessing the trained model performance

```python
# we will use a confusion matrix
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#PREDICT the test results
y_predict_test=NB_classifier.predict(X_test)
cm=confusion_matrix(Y_test,y_predict_test)
sns.heatmap(cm,annot=True)
plt.show()
```

```python
print(classification_report(Y_test,y_predict_test))
```
